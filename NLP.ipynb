```python
# Define stop words paths
stopwords_paths = [
    "/content/StopWords_Currencies.txt",
    "/content/StopWords_Auditor.txt",
    "/content/StopWords_DatesandNumbers.txt",
    "/content/StopWords_Generic.txt",
    "/content/StopWords_GenericLong.txt",
    "/content/StopWords_Geographic.txt",
    "/content/StopWords_Names.txt",

]

# Load stop words from multiple files
for path in stopwords_paths:
    try:
        with open(path, 'r', encoding='utf-8') as file:
            stop_words.update(file.read().splitlines())
    except UnicodeDecodeError:
        # Try a different encoding if utf-8 fails
        try:
            with open(path, 'r', encoding='latin-1') as file:
                stop_words.update(file.read().splitlines())
        except Exception as e:
            print(f"Error reading file {path}: {e}")

# Define positive and negative words paths
positive_words_path = "/positive-words.txt"
negative_words_path = "/negative-words.txt"

# Load positive and negative words
positive_words = set(open(positive_words_path, encoding='utf-8').read().splitlines())
negative_words = set(open(negative_words_path, encoding='latin-1').read().splitlines())




# Inside the clean_text function:
def clean_text(text):
    # Tokenize and remove stop words
    words = [word.lower() for word in word_tokenize(text) if word.isalnum() and word.lower() not in stop_words]

    # Calculate derived variables
    positive_score = sum(1 for word in words if word in positive_words)
    negative_score = sum(1 for word in words if word in negative_words) * -1
    polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)
    subjectivity_score = (positive_score + negative_score) / (len(words) + 0.000001)

    return {
        'positive_score': positive_score,
        'negative_score': negative_score,
        'polarity_score': polarity_score,
        'subjectivity_score': subjectivity_score,
        'cleaned_words': words
    }

def analyze_readability(text):
    # Tokenize sentences
    sentences = sent_tokenize(text)

    # Calculate readability
    average_sentence_length = len(word_tokenize(text)) / len(sentences)
    complex_word_count = sum(1 for word in word_tokenize(text) if len(word) > 2)
    percentage_complex_words = complex_word_count / len(word_tokenize(text))
    fog_index = 0.4 * (average_sentence_length + percentage_complex_words)

    return {
        'average_sentence_length': average_sentence_length,
        'percentage_complex_words': percentage_complex_words,
        'fog_index': fog_index
    }

def extract_personal_pronouns(text):

    personal_pronouns = re.findall(r'\b(?:I|we|my|ours|us)\b', text, flags=re.IGNORECASE)
    return len(personal_pronouns)


excel_file_path = "/output_data_with_hyperlinks.xlsx"
df = pd.read_excel(excel_file_path, engine='openpyxl')


output_df = pd.DataFrame(columns=['URL_ID', 'URL', 'POSITIVE SCORE', 'NEGATIVE SCORE', 'POLARITY SCORE',
                                   'SUBJECTIVITY SCORE', 'AVG SENTENCE LENGTH', 'PERCENTAGE OF COMPLEX WORDS',
                                   'FOG INDEX', 'AVG NUMBER OF WORDS PER SENTENCE', 'COMPLEX WORD COUNT',
                                   'WORD COUNT', 'SYLLABLE PER WORD', 'PERSONAL PRONOUNS', 'AVG WORD LENGTH'])

for index, row in df.iterrows():
    url_id = index + 1
    file_path = row['FilePath']


    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()


    cleaned_text_analysis = clean_text(text)
    readability_analysis = analyze_readability(text)
    personal_pronouns_count = extract_personal_pronouns(text)


    avg_word_length = sum(len(word) for word in cleaned_text_analysis['cleaned_words']) / len(
        cleaned_text_analysis['cleaned_words'])
```
